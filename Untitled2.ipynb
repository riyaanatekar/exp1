```python
#create a list of 5 student names display first and last student
#add a new student remove any one student
#sort the list alphabetically


# Create a list of 5 student names
students = ["riya", "vivaan", "rivya", "maitree", "Karan"]

# Display first and last student
print("First student:", students[0])
print("Last student:", students[-1])

# Add a new student
students.append("Priya")

# Remove any one student (example: remove Karan)
students.remove("Karan")

# Sort the list alphabetically
students.sort()

# Display final list
print("Final student list:", students)
```

```python
#create a list of 5 student marks ,display higest and lowest marks ,calculate average marks , sort marks in acsending order
marks = [78, 92, 85, 64, 88]

print("Highest:", max(marks))
print("Lowest:", min(marks))
print("Average:", sum(marks)/5)

marks.sort()
print("Ascending order:", marks)
```
